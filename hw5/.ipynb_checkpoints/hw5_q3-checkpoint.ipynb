```python
from sklearn.datasets import load_digits
from sklearn.linear_model import Perceptron
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from h5_w7_helper_functions import plot_perceptron_boundary, linear_decision_function
```

```python
data = pd.read_csv("h5w7_data.csv")
```

```python
X_train, y_train = data.iloc[:,:2].to_numpy(), data["label"].to_numpy()
```

a) Code the data points

```python
plt.scatter(X_train[np.where((y_train == 1)),0],X_train[np.where((y_train == 1)),1])
plt.scatter(X_train[np.where((y_train == 2)),0],X_train[np.where((y_train == 2)),1])
```

b) Use sklearnâ€™s implementation of the perceptron to train a perceptron on this dataset.
Obtain and report the classification accuracy

```python
model_perceptron = Perceptron(tol=1e-3, random_state=0, fit_intercept=False)
model_perceptron.fit(X_train,y_train)
```

```python
model_perceptron.score(X_train, y_train)
```

c) Plot the learned decision boundaries from the perceptron in item (b)

```python
weights = model_perceptron.coef_[0]
plot_perceptron_boundary(X_train, y_train, weights, linear_decision_function)
```

d) try to use a quadratic feature space expansion to improve the result

```python
def nonlinear_mapping(X_train):
    return np.concatenate((X_train,\
                        np.reshape(X_train[:,0]*X_train[:,1],(100,1)),\
                        np.reshape(np.power(X_train[:,0],2),(100,1)),\
                        np.reshape(np.power(X_train[:,1],2),(100,1))),axis=1)
```

```python
mapped_X_train = nonlinear_mapping(X_train)

model_perceptron = Perceptron(tol=1e-3, random_state=0, fit_intercept=False)
model_perceptron.fit(mapped_X_train,y_train)
```

```python
model_perceptron.score(mapped_X_train, y_train)
```
